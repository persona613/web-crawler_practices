### 靜態網頁 requests
**台灣水庫即時水情**  \
https://water.taiwanstat.com/

```python
import requests
url = 'https://water.taiwanstat.com/'
res = requests.get(url)
print(res)
print(type(res))
type(res)
```

```python
print(res.history)
print(res.encoding)
print(res.headers)
print(res.cookies)
```

```python
print(res.json()) # 讀取json檔,輸出dict
```

```python
print(res.raw) # bytes stream
```

```python
print(res.content) # bytes
```

```python
print(res.text) # str
```
